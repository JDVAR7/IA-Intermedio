### Suma de 4x4

```python
data = {
 'Nombre': ['Juan', 'Ana', 'Luis', 'María', 'Pedro', 'Camilo', 'Sofia'],
 'Edad': [28, 34, 45, 23, 37, 25, 29],
 'Ingresos_mensuales': [1500, 2000, 2500, 1800, 2200, 1800, 2100]
}
```

```python
# Crear el DataFrame a partir del diccionario
df = pd.DataFrame(data)
print('-------------------------------')
print("Primeras 5 filas del DataFrame:")
print('-------------------------------')
print(df.head(7))
print('__________________________________')
```

```python
# Filtrar las filas donde la columna 'Edad' sea mayor a 30
filtro = df[df['Edad'] > 30]
print('----------------------------------')
print("Filas donde la Edad es mayor a 30:")
print('----------------------------------')
print(filtro)
print('__________________________________')
```

```python
# Crear una nueva columna 'Ingresos_anuales' a partir de
'Ingresos_mensuales'
df['Ingresos_anuales'] = df['Ingresos_mensuales'] * 12
print('-------------------------------')
print("DataFrame con la nueva columna:")
print('-------------------------------')
print(df.head(7))
print('_______________________________________________________')
```

# Pandas

```python
# Mostrar estadísticas descriptivas del DataFrame
print("Estadísticas descriptivas:")
print('______________________________________________________')
print(df.describe())
print('______________________________________________________')

# Visualizar la distribución de una columna con un histograma
import matplotlib.pyplot as plt
df['Edad'].hist(bins=10)
plt.xlabel('Edad', color='purple')
plt.ylabel('Frecuencia', color='purple')
plt.title('Distribución de la Edad', color='purple')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# Definir variables predictoras y objetivo
X = df[['Ingresos_mensuales']]
y = df['Edad']
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
random_state=43)
# Crear el modelo de regresión lineal y entrenarlo
modelo = LinearRegression()
modelo.fit(X_train, y_train)
# Realizar predicciones y calcular el error cuadrático medio
y_pred = modelo.predict(X_test)
error = mean_squared_error(y_test, y_pred)
print("Error cuadrático medio:", error)
```

1. → Precisión
    Porcentaje de predicciones correctas.

2. → Recall
    Porcentaje de casos positivos correctamente identificados.

3. → F1 Score
    Mide el equilibrio entre precisión y recall.

4. → AUC
    Área bajo la curva ROC.

### Agrupación (Clustering) con K-Means

1. → Objetivo

  Agrupar datos similares.

2. → Inicialización

    Elegir puntos iniciales de los clusters.

3. → Asignación

          Asignar cada punto al cluster más cercano.

4. → Actualización

                Recalcular los centroides de los clusters.

5. → Repetición

                       Repetir los pasos 3 y 4 hasta la convergencia.

### Problemas Comunes de ajuste:

Modelo demasiado complejo. ↓

Memoriza datos de entrenamiento. ↓

No generaliza a nuevos datos. Underfitting


Modelo demasiado simple. ↓

No captura patrones importantes. ↓

Rendimiento deficiente en datos de prueba. Overfitting

### Métricas Derivadas de la Matriz de Confusión

#### Exactitud

Porcentaje de predicciones correctas.


#### Precisión
Porcentaje de predicciones positivas correctas.


#### Recall
Porcentaje de casos positivos correctamente identificados.


#### F1 Score
Mide el equilibrio entre precisión y recall.

### Curva ROC y AUC: Evaluando el Rendimiento del Modelo


#### Curva ROC
Representa la relación entre la tasa de verdaderos positivos y la tasa de falsos positivos.


#### AUC
Área bajo la curva ROC, mide la capacidad del modelo para distinguir entre clases

### Técnicas para Mitigar Overfitting

#### Regularización

✔Reduce la complejidad del modelo.

✔Evita que el modelo se ajuste demasiado a los datos de entrenamiento.


### Aumento de Datos

✔Crea nuevas muestras de datos a partir de los existentes.

✔Mejora la generalización del modelo.

### Técnicas para Mitigar Underfitting

#### Validación Cruzada

Divide los datos en varios conjuntos.

Entrena y evalúa el modelo en cada conjunto.

#### Selección de Atributos

Selecciona las variables más relevantes.

Elimina variables irrelevantes o redundantes.
