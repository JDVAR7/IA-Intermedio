```python
import sqlite3

conexion = sqlite3.connect('laboratorio.db')
cursor = conexion.cursor()

# Crear tabla Inscripciones con FOREIGN KEYS
cursor.execute("""
    CREATE TABLE IF NOT EXISTS Inscripciones (
        InscripcionID INTEGER PRIMARY KEY,
        EstudianteID  INTEGER,
        CursoID       INTEGER,
        FOREIGN KEY (EstudianteID) REFERENCES Estudiantes(ID),
        FOREIGN KEY (CursoID)      REFERENCES Cursos(CursoID)
    )
""")

# Insertar inscripciones (relación estudiante-curso)
cursor.executemany(
    'INSERT OR IGNORE INTO Inscripciones (InscripcionID, EstudianteID, CursoID) VALUES (?,?,?)',
    [
        (1, 1, 101),  # Laura Gómez → Cálculo
        (2, 2, 102),  # Juan Pérez  → Programación
        (3, 3, 103),  # Ana Torres  → Física General
        (4, 4, 104),  # Maria tavarez → Integrales
        (5, 5, 105),  # David galindo → Programación
        (6, 6, 106),  # Claudia tovar → Física Cuantica
        (7, 7, 107),  # Esther escobar → hilos y agujas
        (8, 8, 108),  # Miguel abdulá → reino vegetal
        (9, 9, 109)   # Diego alvaréz → Materialismo
    ]
)
conexion.commit()

# Consulta
# JOIN: combinar las 9 tablas
sql_join = """
    SELECT
        Estudiantes.Nombre  AS Estudiante,
        Cursos.Nombre       AS Curso,
        Cursos.Creditos
    FROM Inscripciones
    JOIN Estudiantes ON Inscripciones.EstudianteID = Estudiantes.ID
    JOIN Cursos      ON Inscripciones.CursoID      = Cursos.CursoID
"""
cursor.execute(sql_join)
resultados = cursor.fetchall()

print('Resultados del JOIN:')
for fila in resultados:
    print(f'  Estudiante: {fila[0]}, Curso: {fila[1]}, Créditos: {fila[2]}')

conexion.commit()
conexion.close()
```

```python
import sqlite3

# ══════════════════════════════════════════════════════════════
# RETO FINAL — SISTEMA ACADÉMICO COMPLETO
# Sesión 19 — Laboratorio 19 — TalentoTech IA
# ══════════════════════════════════════════════════════════════

conexion = sqlite3.connect(':memory:')  # En Colab usa ':memory:' para pruebas
cursor = conexion.cursor()

# ── 1. CREAR TODAS LAS TABLAS ───────────────────────────────
cursor.executescript('''
    CREATE TABLE IF NOT EXISTS Estudiantes (
        ID      INTEGER PRIMARY KEY AUTOINCREMENT,
        Nombre  TEXT    NOT NULL,
        Edad    INTEGER NOT NULL,
        Carrera TEXT    NOT NULL
    );
    CREATE TABLE IF NOT EXISTS Cursos (
        CursoID  INTEGER PRIMARY KEY AUTOINCREMENT,
        Nombre   TEXT    NOT NULL,
        Creditos INTEGER NOT NULL
    );
    CREATE TABLE IF NOT EXISTS Inscripciones (
        InscripcionID INTEGER PRIMARY KEY AUTOINCREMENT,
        EstudianteID  INTEGER REFERENCES Estudiantes(ID),
        CursoID       INTEGER REFERENCES Cursos(CursoID)
    );
    CREATE TABLE IF NOT EXISTS Calificaciones (
        CalifID      INTEGER PRIMARY KEY AUTOINCREMENT,
        EstudianteID INTEGER REFERENCES Estudiantes(ID),
        CursoID      INTEGER REFERENCES Cursos(CursoID),
        Nota         REAL    NOT NULL CHECK(Nota>=0.0 AND Nota<=5.0)
    );
''')
print('✅ Tablas creadas: Estudiantes, Cursos, Inscripciones, Calificaciones')

# ── 2. INSERTAR DATOS ───────────────────────────────────────
estudiantes = [
    ('Laura Gómez',  22, 'Ingeniería'),
    ('Juan Pérez',   24, 'Matemáticas'),
    ('Ana Torres',   21, 'Física'),
    ('Carlos Ruiz',  23, 'Sistemas'),
    ('María López',  20, 'Ingeniería'),
]
cursor.executemany(
    'INSERT INTO Estudiantes (Nombre,Edad,Carrera) VALUES (?,?,?)',
    estudiantes
)

cursos = [
    ('Cálculo',         4),
    ('Programación',    3),
    ('Física General',  4),
    ('Estadística',     3),
]
cursor.executemany(
    'INSERT INTO Cursos (Nombre,Creditos) VALUES (?,?)',
    cursos
)

inscripciones = [
    (1,1),(1,2),(1,4),  # Laura: Cálculo, Prog, Estadística
    (2,1),(2,3),        # Juan: Cálculo, Física
    (3,2),(3,3),        # Ana: Prog, Física
    (4,1),(4,4),        # Carlos: Cálculo, Estadística
    (5,2),(5,3),(5,4),  # María: Prog, Física, Estadística
]
cursor.executemany(
    'INSERT INTO Inscripciones (EstudianteID,CursoID) VALUES (?,?)',
    inscripciones
)

calificaciones = [
    (1,1,4.5),(1,2,4.8),(1,4,4.2),  # Laura
    (2,1,3.8),(2,3,4.1),             # Juan
    (3,2,3.5),(3,3,3.9),             # Ana
    (4,1,4.7),(4,4,4.9),             # Carlos
    (5,2,3.2),(5,3,3.6),(5,4,3.0),  # María
]
cursor.executemany(
    'INSERT INTO Calificaciones (EstudianteID,CursoID,Nota) VALUES (?,?,?)',
    calificaciones
)
conexion.commit()
print('✅ Datos insertados correctamente.')

# ── 3. CONSULTAS DE ANÁLISIS ────────────────────────────────

# PREGUNTA 1: ¿Cuántos estudiantes están inscritos en cada curso?
print('\n📊 PREGUNTA 1: Estudiantes por curso')
print('-'*45)
cursor.execute('''
    SELECT Cursos.Nombre, COUNT(Inscripciones.EstudianteID) AS Total
    FROM Cursos
    LEFT JOIN Inscripciones ON Cursos.CursoID = Inscripciones.CursoID
    GROUP BY Cursos.Nombre
    ORDER BY Total DESC
''')
for row in cursor.fetchall():
    print(f'  {row[0]:<20} → {row[1]} estudiantes')

# PREGUNTA 2: ¿Cuál es la nota promedio por curso?
print('\n📊 PREGUNTA 2: Nota promedio por curso')
print('-'*45)
cursor.execute('''
    SELECT Cursos.Nombre, ROUND(AVG(Calificaciones.Nota),2) AS Promedio
    FROM Calificaciones
    JOIN Cursos ON Calificaciones.CursoID = Cursos.CursoID
    GROUP BY Cursos.Nombre
    ORDER BY Promedio DESC
''')
for row in cursor.fetchall():
    print(f'  {row[0]:<20} → Promedio: {row[1]}')

# PREGUNTA 3: ¿Qué estudiantes tienen promedio mayor a 4.0?
print('\n📊 PREGUNTA 3: Estudiantes con promedio > 4.0')
print('-'*45)
cursor.execute('''
    SELECT Estudiantes.Nombre, ROUND(AVG(Calificaciones.Nota),2) AS Promedio
    FROM Calificaciones
    JOIN Estudiantes ON Calificaciones.EstudianteID = Estudiantes.ID
    GROUP BY Estudiantes.Nombre
    HAVING AVG(Calificaciones.Nota) > 4.0
    ORDER BY Promedio DESC
''')
resultados = cursor.fetchall()
if resultados:
    for row in resultados:
        print(f'  ⭐ {row[0]:<20} → Promedio: {row[1]}')
else:
    print('  Ningún estudiante supera el promedio de 4.0')

conexion.close()
print('\n🏁 Reto Final completado exitosamente.')
```

```python
# ── EJERCICIO 1 CORREGIDO — CRUD COMPLETO EN MONGODB ───────────
# CORRECCIÓN: Se usa mongomock en lugar de MongoClient real
# Esto permite ejecutar en Google Colab sin servidor MongoDB instalado

import mongomock                          # ← CORRECCIÓN: simula MongoDB en RAM
from bson import ObjectId
import json

# PASO 1: Conectar usando el cliente simulado
client = mongomock.MongoClient()         # ← Sin URL: crea BD en memoria
print('✅ Conexión simulada a MongoDB establecida.')

# PASO 2: Crear base de datos y colección
db = client['estudiantes']               # Crear/conectar BD 'estudiantes'
collection = db['informacion']           # Crear/conectar colección 'informacion'
print('✅ Base de datos y colección seleccionadas.')

# PASO 3: Insertar UN documento (insert_one)
estudiante = {
    'nombre': 'Laura Sánchez',
    'edad': 22,
    'carrera': 'Ingeniería de Sistemas',
    'promedio': 4.3
}
resultado = collection.insert_one(estudiante)
print(f'✅ Documento insertado con ID: {resultado.inserted_id}')

# PASO 4: Insertar MÚLTIPLES documentos (insert_many)
estudiantes = [
    {'nombre': 'Jorge Morales', 'edad': 24, 'carrera': 'Matemáticas', 'promedio': 4.1},
    {'nombre': 'Liliana Gómez', 'edad': 21, 'carrera': 'Física',       'promedio': 4.7}
]
resultado_many = collection.insert_many(estudiantes)
print(f'✅ Múltiples documentos insertados: {len(resultado_many.inserted_ids)} documentos')

# PASO 5: LEER — consultar todos los documentos
print('\n📋 TODOS LOS ESTUDIANTES:')
for doc in collection.find():
    print(f'  → {doc["nombre"]} | {doc["carrera"]} | Promedio: {doc["promedio"]}')

# PASO 6: ACTUALIZAR — cambiar el promedio de Laura
collection.update_one(
    {'nombre': 'Laura Sánchez'},
    {'$set': {'promedio': 4.5}}          # $set actualiza solo ese campo
)
print('\n✏️ Promedio de Laura actualizado a 4.5')

# PASO 7: ELIMINAR — borrar a Liliana Gómez
collection.delete_one({'nombre': 'Liliana Gómez'})
print('🗑️ Liliana Gómez eliminada de la colección')

# PASO 8: Verificar estado final
print('\n📋 ESTADO FINAL DE LA COLECCIÓN:')
for doc in collection.find():
    print(f'  → {doc["nombre"]} | {doc["carrera"]} | Promedio: {doc["promedio"]}')
print(f'Total documentos: {collection.count_documents({})}')
```

```python
# ── EJERCICIO 2 CORREGIDO — OPERACIONES BÁSICAS REDIS ──────────
# CORRECCIÓN: Se usa fakeredis en lugar de redis.StrictRedis

import fakeredis                          # ← CORRECCIÓN: simula Redis en RAM

# PASO 1: Crear cliente Redis simulado
client = fakeredis.FakeRedis(decode_responses=True)
# decode_responses=True: los valores se devuelven como strings, no bytes
print('✅ Conexión simulada a Redis establecida.')

# PASO 2: SET — guardar clave-valor simple
client.set('usuario:1001', 'Laura Sánchez')   # clave → valor
print('✅ Clave usuario:1001 almacenada.')

# PASO 3: GET — recuperar valor por clave
usuario = client.get('usuario:1001')
print(f'Valor recuperado: {usuario}')          # Output: Laura Sánchez

# PASO 4: SETEX — clave con tiempo de vida (TTL)
client.setex('token:123', 300, 'activo')
# setex(clave, segundos, valor): 300 seg = 5 minutos
print('✅ Token con TTL de 300 segundos creado.')

# Verificar TTL restante
ttl_restante = client.ttl('token:123')
print(f'Tiempo de vida restante del token: {ttl_restante} segundos')

# PASO 5: RPUSH — agregar elementos a una lista
client.rpush('tareas', 'Tarea1', 'Tarea2', 'Tarea3')
# rpush: agrega al FINAL de la lista (Right PUSH)
print('✅ Tareas almacenadas en lista.')

# PASO 6: LRANGE — obtener rango de la lista
tareas = client.lrange('tareas', 0, -1)
# lrange(clave, inicio, fin): 0 = primero, -1 = último
print(f'Tareas en la lista: {tareas}')

# Verificar número de elementos en la lista
print(f'Total de tareas: {client.llen("tareas")}')
```

```python
# ── EJERCICIO 3 CORREGIDO — ESCALABILIDAD Y TTL EN REDIS ───────
import fakeredis
import time

client = fakeredis.FakeRedis(decode_responses=True)

# PARTE 1: Crear múltiples claves con TTL diferentes
print('='*55)
print('PARTE 1: Creando claves con TTL progresivos')
print('='*55)

for i in range(5):
    ttl_segundos = 10 * (i + 1)          # 10, 20, 30, 40, 50 segundos
    valor = f'valor_{i}'
    client.setex(f'clave:{i}', ttl_segundos, valor)
    print(f'✅ clave:{i} = {valor!r} | TTL: {ttl_segundos}s')

# PARTE 2: Verificar TTL de todas las claves
print('\n' + '='*55)
print('PARTE 2: Verificando tiempos de vida')
print('='*55)

for i in range(5):
    ttl = client.ttl(f'clave:{i}')
    valor = client.get(f'clave:{i}')
    print(f'clave:{i} → valor={valor!r}, TTL restante={ttl}s')

# PARTE 3: Verificar clave sin TTL
client.set('clave_permanente', 'nunca_expira')
ttl_perm = client.ttl('clave_permanente')
print(f'\nclave_permanente → TTL: {ttl_perm} (-1 = sin expiración)')

# PARTE 4: Simular expiración manual
client.expire('clave:0', 1)               # Forzar TTL a 1 segundo
time.sleep(1.1)                            # Esperar que expire
valor_expirado = client.get('clave:0')
print(f'\nclave:0 después de expirar: {valor_expirado} (None = expirada)')

# PARTE 5: Resumen de claves activas
print('\n' + '='*55)
print('RESUMEN: Claves activas en Redis')
print('='*55)
claves = client.keys('*')                 # Listar todas las claves
print(f'Total claves activas: {len(claves)}')
for clave in sorted(claves):
    ttl = client.ttl(clave)
    val = client.get(clave)
    ttl_str = f'{ttl}s' if ttl > 0 else 'permanente'
    print(f'  {clave:<20} → {str(val):<15} | TTL: {ttl_str}')
```

```python
# ── EJERCICIO 4 CORREGIDO — DATOS JERÁRQUICOS EN MONGODB ───────
import mongomock

# CORRECCIÓN: Nueva conexión para este ejercicio
client_mongo = mongomock.MongoClient()  # ← Variable renombrada para no pisar client Redis

db = client_mongo['jerarquia']
collection = db['organizacion']
print('✅ BD jerarquia y colección organizacion listas.')

# PASO 1: Insertar documento jerárquico con subdocumentos anidados
organizacion = {
    'departamento': 'Tecnología',
    'jefe': 'Laura Sánchez',
    'empleados': [                        # ← Array de subdocumentos embebidos
        {'nombre': 'Jorge Morales', 'rol': 'Desarrollador'},
        {'nombre': 'Liliana Gómez', 'rol': 'Analista de Datos'},
        {'nombre': 'Carlos Ruiz',   'rol': 'Soporte Técnico'}
    ]
}
resultado = collection.insert_one(organizacion)
print(f'✅ Documento jerárquico insertado. ID: {resultado.inserted_id}')

# PASO 2: Consultar todos los documentos
print('\n📋 DOCUMENTOS EN LA COLECCIÓN:')
for doc in collection.find():
    print(f'  Dpto: {doc["departamento"]} | Jefe: {doc["jefe"]}')
    for emp in doc.get('empleados', []):
        print(f'    └─ {emp["nombre"]} → {emp["rol"]}')

# PASO 3: Consultar subdocumento específico por campo anidado
# Notación punto: 'empleados.nombre' busca DENTRO del array
# CORRECCIÓN: mongomock no soporta la proyección posicional '$'.
# Recuperamos el documento completo y filtramos en Python.
empleado_documento = collection.find_one(
    {'empleados.nombre': 'Liliana Gómez'}
)

empleado_encontrado = None
if empleado_documento and 'empleados' in empleado_documento:
    for emp in empleado_documento['empleados']:
        if emp.get('nombre') == 'Liliana Gómez':
            empleado_encontrado = emp
            break

print(f'\n🔍 Subdocumento encontrado: {empleado_encontrado}')

# PASO 4: Actualizar subdocumento con el posicional $
collection.update_one(
    {'empleados.nombre': 'Carlos Ruiz'},  # Filtro: doc que contiene ese empleado
    {'$set': {'empleados.$.rol': 'Administrador de Sistemas'}}  # $ = índice encontrado
)
print('\n✏️ Rol de Carlos Ruiz actualizado a Administrador de Sistemas.')

# PASO 5: Eliminar subdocumento usando $pull
collection.update_one(
    {'departamento': 'Tecnología'},
    {'$pull': {'empleados': {'nombre': 'Jorge Morales'}}}  # $pull elimina del array
)
print('🗑️ Jorge Morales eliminado del array empleados.')

# PASO 6: Verificar documento actualizado
doc_final = collection.find_one({'departamento': 'Tecnología'})
print('\n📋 DOCUMENTO FINAL:')
print(f'  Departamento: {doc_final["departamento"]} | Jefe: {doc_final["jefe"]}')
print('  Empleados actuales:')
for emp in doc_final.get('empleados', []):
    print(f'    └─ {emp["nombre"]} → {emp["rol"]}')
```

```python
# ══════════════════════════════════════════════════════════════
# RETO FINAL — SISTEMA DE GESTIÓN DE USUARIOS
# MongoDB: perfiles permanentes + Redis: sesiones temporales
# Sesión 20 — TalentoTech IA Intermedio
# ══════════════════════════════════════════════════════════════

import mongomock
import fakeredis
import uuid
import time
import hashlib
from datetime import datetime

# ── CONEXIONES ─────────────────────────────────────────────
mongo_client = mongomock.MongoClient()
redis_client = fakeredis.FakeRedis(decode_responses=True)

db = mongo_client['app_usuarios']
usuarios_col = db['usuarios']

print('✅ MongoDB y Redis conectados (modo simulación Colab)')

# ── FUNCIONES DEL SISTEMA ───────────────────────────────────

def registrar_usuario(nombre, email, rol='estudiante'):
    '''Registra un usuario en MongoDB'''
    # Verificar si ya existe
    if usuarios_col.find_one({'email': email}):
        return None, 'Email ya registrado'
    usuario = {
        'nombre': nombre,
        'email': email,
        'rol': rol,
        'creado_en': datetime.now().isoformat(),
        'activo': True,
        'historial_sesiones': []
    }
    resultado = usuarios_col.insert_one(usuario)
    return str(resultado.inserted_id), 'Usuario registrado'

def iniciar_sesion(email):
    '''Crea token de sesión en Redis con TTL'''
    usuario = usuarios_col.find_one({'email': email})
    if not usuario:
        return None, 'Usuario no encontrado'
    # Generar token único
    token = str(uuid.uuid4())[:16]
    user_id = str(usuario['_id'])
    # Guardar token en Redis con TTL de 1800 segundos (30 minutos)
    redis_client.setex(f'session:{token}', 1800, user_id)
    print(f'  Token: {token} | TTL: 1800s | UserID: {user_id}')
    return token, 'Sesión iniciada'

def validar_sesion(token):
    '''Valida si el token de sesión es activo en Redis'''
    user_id = redis_client.get(f'session:{token}')
    if not user_id:
        return None, 'Sesión expirada o inválida'
    ttl_restante = redis_client.ttl(f'session:{token}')
    return user_id, f'Sesión válida ({ttl_restante}s restantes)'

def obtener_perfil(user_id):
    '''Recupera el perfil completo desde MongoDB'''
    from bson import ObjectId
    try:
        usuario = usuarios_col.find_one({'_id': ObjectId(user_id)})
        if usuario:
            usuario['_id'] = str(usuario['_id'])   # Convertir para visualizar
        return usuario
    except Exception:
        return None

def cerrar_sesion(token):
    '''Elimina el token de Redis inmediatamente'''
    eliminados = redis_client.delete(f'session:{token}')
    return eliminados > 0

# ── DEMO DEL SISTEMA ────────────────────────────────────────
print('\n' + '='*55)
print('PASO 1: REGISTRAR USUARIOS EN MONGODB')
print('='*55)

usuarios_demo = [
    ('Laura Sánchez', 'laura@talento.com', 'estudiante'),
    ('Jorge Morales',  'jorge@talento.com', 'estudiante'),
    ('Admin Sistema',  'admin@talento.com', 'admin'),
]

for nombre, email, rol in usuarios_demo:
    uid, msg = registrar_usuario(nombre, email, rol)
    print(f'  {nombre} → {msg} | ID: {uid[:8] if uid else "N/A"}...')

print(f'\nTotal usuarios en MongoDB: {usuarios_col.count_documents({})}')

print('\n' + '='*55)
print('PASO 2: INICIAR SESIONES (TOKENS EN REDIS)')
print('='*55)

token_laura, msg1 = iniciar_sesion('laura@talento.com')
token_jorge, msg2 = iniciar_sesion('jorge@talento.com')
print(f'Laura: {msg1}')
print(f'Jorge: {msg2}')

print('\n' + '='*55)
print('PASO 3: VALIDAR SESIONES Y OBTENER PERFILES')
print('='*55)

user_id, status = validar_sesion(token_laura)
print(f'Validación token Laura: {status}')
if user_id:
    perfil = obtener_perfil(user_id)
    if perfil:
        print(f'  Perfil: {perfil["nombre"]} | Rol: {perfil["rol"]}')

print('\n' + '='*55)
print('PASO 4: SESIÓN INVÁLIDA (TOKEN EXPIRADO)')
print('='*55)

cerrar_sesion(token_jorge)                 # Simular cierre de sesión
_, status_inv = validar_sesion(token_jorge)
print(f'Token Jorge tras cerrar sesión: {status_inv}')

print('\n' + '='*55)
print('PASO 5: RESUMEN FINAL DEL SISTEMA')
print('='*55)
print(f'Usuarios en MongoDB: {usuarios_col.count_documents({})}')
sesiones_activas = len([k for k in redis_client.keys('session:*')])
print(f'Sesiones activas en Redis: {sesiones_activas}')
print('\n✅ Reto Final completado exitosamente.')
```
